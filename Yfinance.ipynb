```python

import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import pandas as pd

# List of tickers for the bond market
tickers = ["^TNX", "LQD", "HYG", "JNK", "VCIT"]

# Set date range
start_date = datetime.datetime(1990, 1, 1)
end_date = datetime.datetime(2023, 1, 1)

# Download data from Yahoo Finance
data = yf.download(tickers, start=start_date, end=end_date)

# Save data to a CSV file
data.to_csv("bond_market.csv")

# Select index and allocation weight for investment
index_choice = "^TNX"  # You can choose a different index here
allocation_weight = 0.3  # You can specify a different weight (from 0 to 1) here

# Calculate returns for the selected index
data["Returns"] = data["Close"][index_choice].pct_change()
data["Cumulative Returns"] = (1 + data["Returns"]).cumprod()

# Calculate returns with allocation
data["Allocation"] = data["Cumulative Returns"] * allocation_weight

# Plot returns with allocation
data[["Cumulative Returns", "Allocation"]].plot(figsize=(10, 6))
plt.title("Returns with Allocation")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(["Without Allocation", "With Allocation"])
plt.show()

# Display table with weights and returns
weights = pd.DataFrame(data={"Index": [index_choice], "Weight": [allocation_weight]})
returns = pd.DataFrame(data={"Date": data.index, "Returns": data["Returns"]})

print("Table with Weights:")
print(weights)
print("\nTable with Returns:")
print(returns)
```
