```python

import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import pandas as pd

# Список тикеров для топ 15 индексов
tickers = ["^GSPC", "^DJI", "^IXIC", "^RUT", "^VIX", "^TNX", "GC=F", "SI=F", "CL=F", "^IRX", "^FVX", "^TYX", "^US10Y", "^US20Y", "^US30Y"]

# Установка диапазона дат
start_date = datetime.datetime(1990, 1, 1)
end_date = datetime.datetime(2023, 1, 1)

# Загрузка данных с Yahoo Finance
data = yf.download(tickers, start=start_date, end=end_date)

# Сохранение данных в CSV файл
data.to_csv("индексы.csv")

# Выбор индекса и веса для инвестирования
index_choice = "^GSPC"  # Здесь можно выбрать другой индекс
allocation_weight = 0.2  # Здесь можно указать другой вес (от 0 до 1)

# Рассчитываем доходность для выбранного индекса
data["Returns"] = data["Close"][index_choice].pct_change()
data["Cumulative Returns"] = (1 + data["Returns"]).cumprod()

# Рассчитываем доходность с учетом аллокации средств
data["Allocation"] = data["Cumulative Returns"] * allocation_weight

# Построение графика доходности с учетом аллокации
data[["Cumulative Returns", "Allocation"]].plot(figsize=(10, 6))
plt.title("Доходность с учетом аллокации")
plt.xlabel("Дата")
plt.ylabel("Доходность")
plt.legend(["Без аллокации", "С аллокацией"])
plt.show()

# Вывод таблицы с весами и доходностью
weights = pd.DataFrame(data={"Индекс": [index_choice], "Вес": [allocation_weight]})
returns = pd.DataFrame(data={"Дата": data.index, "Доходность": data["Returns"]})
print("Таблица с весами:")
print(weights)
print("\nТаблица с доходностью:")
print(returns)
```

```python

import yfinance as yf
import datetime
import matplotlib.pyplot as plt
import pandas as pd

# List of tickers for the bond market
tickers = ["^TNX", "LQD", "HYG", "JNK", "VCIT"]

# Set date range
start_date = datetime.datetime(1990, 1, 1)
end_date = datetime.datetime(2023, 1, 1)

# Download data from Yahoo Finance
data = yf.download(tickers, start=start_date, end=end_date)

# Save data to a CSV file
data.to_csv("bond_market.csv")

# Select index and allocation weight for investment
index_choice = "^TNX"  # You can choose a different index here
allocation_weight = 0.3  # You can specify a different weight (from 0 to 1) here

# Calculate returns for the selected index
data["Returns"] = data["Close"][index_choice].pct_change()
data["Cumulative Returns"] = (1 + data["Returns"]).cumprod()

# Calculate returns with allocation
data["Allocation"] = data["Cumulative Returns"] * allocation_weight

# Plot returns with allocation
data[["Cumulative Returns", "Allocation"]].plot(figsize=(10, 6))
plt.title("Returns with Allocation")
plt.xlabel("Date")
plt.ylabel("Returns")
plt.legend(["Without Allocation", "With Allocation"])
plt.show()

# Display table with weights and returns
weights = pd.DataFrame(data={"Index": [index_choice], "Weight": [allocation_weight]})
returns = pd.DataFrame(data={"Date": data.index, "Returns": data["Returns"]})

print("Table with Weights:")
print(weights)
print("\nTable with Returns:")
print(returns)
```

```python

import numpy as np

# Задайте данные для расчетов
диапазон_доходности_классов_активов = [0.05, 0.08, 0.1]
диапазон_весов_классов_активов = [0.3, 0.4, 0.3]

диапазон_доходности_инструментов = [0.07, 0.09, 0.11]
диапазон_весов_инструментов = [0.25, 0.35, 0.4]
волатильность_инструмента = 0.05

вес_актива = 0.3
увеличение_ВВП = 0.02
доля_акций = 0.4
увеличение_ключевой_ставки = 0.01
доходность_облигаций = 0.05
снижение_ставки = 0.02
доля_металлов = 0.1
рост_безработицы_и_инфляции = 0.03
доля_иностранной_валюты = 0.2
увеличение_доли_производных = 0.05
рост_индекса_волатильности = 0.02

# Расчет средневзвешенных весов на основе доходности классов активов
средневзвешенные_веса_классов_активов = np.average(диапазон_доходности_классов_активов, weights=диапазон_весов_классов_активов)

# Расчет средневзвешенных весов на основе доходности и волатильности инструментов внутри классов активов
средневзвешенные_веса_инструментов = np.average(диапазон_доходности_инструментов, weights=диапазон_весов_инструментов) * (1 - волатильность_инструмента)

# Формула изменения весов активов с учетом изменения факторов влияющих на каждый класс активов
изменение_весов_активов = вес_актива + (увеличение_ВВП * доля_акций) + (увеличение_ключевой_ставки * доходность_облигаций) - (снижение_ставки * доля_акций) + (рост_безработицы_и_инфляции * доля_металлов) - (уменьшение_ВВП * доля_иностранной_валюты) + (увеличение_доли_производных * рост_индекса_волатильности)

# Вывод результатов
print("Средневзвешенные веса на основе доходности классов активов:", средневзвешенные_веса_классов_активов)
print("Средневзвешенные веса на основе доходности и волатильности инструментов:", средневзвешенные_веса_инструментов)
print("Изменение весов активов с учетом факторов:", изменение_весов_активов)
```
